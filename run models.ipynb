```python
from mliv.inference import Vanilla2SLS
from mliv.inference import NN2SLS
from mliv.utils import CausalDataset
```

```python
data = CausalDataset('./basic/')

model = NN2SLS()
model.config['device'] = None
model.config['epochs'] = 20
model.fit(data)
a1, a2= model.ATE(data.train)

print(f"NN: {a1}, {a2}")
```

```python
data = CausalDataset('./basic/')
model = Vanilla2SLS()
model.fit(data)
a1, a2 = model.ATE(data.train)

print(f"Vanilla: {a1}, {a2}")

model = NN2SLS()
model.config['device'] = None
model.config['epochs'] = 20
model.fit(data)
a1, a2= model.ATE(data.train)

print(f"NN: {a1}, {a2}")
```
